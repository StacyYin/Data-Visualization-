```python
# plotly imports
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode

# plotly + cufflinks
import cufflinks
cufflinks.go_offline(connected = True)
cufflinks.set_config_file(world_readable = True, theme = 'pearl')
init_notebook_mode(connected = True)

import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
df = pd.read_csv('Chile.csv')
df.head()
```

## Histograms

```python
df['income'].iplot(
kind = 'hist',
bins = 30,
xTitle = 'income',
linecolor = 'black',
yTitle = 'amount',
title = 'Income Distribution')
```

```python
# Using percentage
df['income'].iplot(
kind = 'hist',
bins = 30,
histnorm = 'percent',
xTitle = 'income',
linecolor = 'black',
yTitle = 'percentage (%)',
title = 'Income Distribution in Percent')
```

```python
df[['population','income']].iplot(
kind = 'hist',
bins = 20,
linecolor = 'black',
histnorm = 'percent',
bargap = 0.1,
opacity = 0.9,
barmode = 'group',
yTitle = 'percentage (%)',
title = 'income and population'
)
```

```python
# Overlaid graph
df[['population','income']].iplot(
kind = 'hist',
bins = 20,
linecolor = 'black',
histnorm = 'percent',
bargap = 0.1,
opacity = 0.6,
barmode = 'overlay',
yTitle = 'percentage (%)',
title = 'income and population'
)
```

```python
# Bar chart
df.groupby('region').mean()['age'].iplot(
kind = 'bar',
yTitle = 'Average Age by Region',
bins = 10,
linecolor = 'black')
```

```python
df.groupby('region').mean()[['population','income']].iplot(
kind = 'bar',
title = 'Average Age and Income by Region',
bargap = .1,
bins = 10,
linecolor = 'black')
```

```python
df[['population','income']].iplot(
kind = 'box',
yTitle = 'population and income distribution')
```

```python
df2 = df.pivot(columns = 'education', values = 'income')
df2 = df2.iloc[:,1:]
df2.head()
```

```python
df2.iplot(
kind = 'box',
layout = dict(
title = 'relationship between education and income',
margin = dict(b = 100)))
```

```python
df.groupby('region').mean()[['income','population']].iplot(
kind = 'spread',
mode = 'lines+markers',
title = 'spread between population and income by region'
)
```
